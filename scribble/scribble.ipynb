```python
def u1(lam):
    return np.matrix([[1,0],[0,np.exp(1j*lam)]])

def u2(phi, lam):
    return np.matrix(
        [
            [1/np.sqrt(2),-np.exp(1j*lam)*1/np.sqrt(2)],
            [np.exp(1j*phi)*1/np.sqrt(2),np.exp(1j*lam+1j*phi)*1/np.sqrt(2)]
        ])

def u3(theta,phi,lam):
    return np.matrix([
        [np.cos(theta/2),-np.exp(1j*lam) * np.sin(theta/2)],
        [np.exp(1j*phi) * np.sin(theta/2),np.exp(1j*lam+1j*phi)*np.cos(theta/2)]
    ])

def roundFloat(matrix: np.ndarray):
    return matrix.round(matrix, 2)
```
